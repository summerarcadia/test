# I. Matplotlib Plotting
* Plotting x and y points
* x-axis == horizontal axis
* y-axis == vertical axis

```python
# Example: Draw a line in a diagram from position (1, 4) to position (6, 8)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 4])
ypoints = np.array([6, 8])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Example: Draw two points in the diagram, one at position (1, 4) and one in position (6, 8)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 4])
ypoints = np.array([6, 8])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

# II. Matplotlib Subplot
* Display Multiple Plots: use *subplot()* function

```python
# Example: Draw 2 plots

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 5, 10, 6])

plt.subplot(1, 2, 1) # this figure has 1 row, 2 columns, and is the first plot
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([1, 3, 5, 7])

plt.subplot(1, 2, 2) # this figure has 1 row, 2 columns, and is the second plot
plt.plot(x,y)

plt.show()
```

```python
# Example: Draw 2 plots on top of each other

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 5, 10, 6])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([1, 3, 5, 7])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

# Title
* Use *title()* function to add a title to each plot

```python
# Example: Draw 2 plots with titles

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 5, 8, 6])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([1, 3, 5, 7])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

# Super Title
* Use *suptitle()* function to add a title to the entire figure

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Figure 1")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([1, 3, 5, 7])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Figure 2")

plt.suptitle("MY PLOT")
plt.show()
```

```python
# Example: Draw 4 bars

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["I", "II", "III", "IV"])
y = np.array([8, 5, 6, 9])

plt.bar(x,y) # 1st and 2nd argument as arrays
plt.show()
```

# Horizontal Bars
* Use *barh()* function to display bars horizontally

```python
# Example: Draw 4 horizontal bars


import matplotlib.pyplot as plt
import numpy as np

x = np.array(["I", "II", "III", "IV"])
y = np.array([8, 5, 6, 9])

plt.barh(x, y)
plt.show()
```

# Matplotlib Pie Charts

# Creating Pie Charts
* Use *pie()* function to draw pie charts
* Pie chart draws one piece (aka wedge) for each value in the array
* By default the plotting of the 1st wedge starts from the x-axis and move *counterclockwise*

```python
# Example: A simple pie chart

import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 25, 30, 30])

plt.pie(y)
plt.show() 
```

# Labels
* Use label parameter to add labels to the pie chart
* The label parameter must be an array with one label for each wedge

```python
# Example: A simple pie chart

import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 25, 30, 30])
mylabels = ["I", "II", "III", "IV"]

plt.pie(y, labels = mylabels)
plt.show() 
```

# Start Angle
* The default start angle is at the x-axis, but can be changed by specifying a startangle parameter
* The startangle parameter is defined with an angle in degrees, default angle is 0

```python
# Example: Start the first wedge at 90 degrees

import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 25, 30, 30])
mylabels = ["I", "II", "III", "IV"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

# Explode
* Use *explode* parameter to make one of the wedges stand out
* The explode parameter must be an array with one value for each wedge, and not None, if specified
* Each value represents how far from the center each wedge is displayed

```python
# Example: Pull the "III" wedge 0.3 from the center of the pie

import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 25, 30, 30])
mylabels = ["I", "II", "III", "IV"]
myexplode = [0, 0, 0.3, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

# Shadow
* Set *shadows* parameter == True to add a shadow to the pie chart

```python
# Example: Add a shadow

import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 25, 30, 30])
mylabels = ["I", "II", "III", "IV"]
myexplode = [0, 0, 0.3, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

# Colors
* Use *colors* parameter to set the color of each wedge
* The colors parameter must be an array with one value for each wedge, if specified
* Shortcuts
  * 'r' - Red
  * 'g' - Green
  * 'b' - Blue
  * 'c' - Cyan
  * 'm' - Magenta
  * 'y' - Yellow
  * 'k' - Black
  * 'w' - White

```python
# Example: Specify a new color for each wedge

import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 25, 30, 30])
mylabels = ["I", "II", "III", "IV"]
mycolors = ["red", "orange", "gray", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

# Legend
* Use *legend()* function to add a list of explanation for each wedge

```python
# Example: Add a legend

import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 25, 30, 30])
mylabels = ["I", "II", "III", "IV"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

# Legend With Header
* Add *title* parameter to the legend function to add a header to the legend

```python
# Example: Add a legend with a header

import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 25, 30, 30])
mylabels = ["I", "II", "III", "IV"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Stages:")
plt.show()
```
